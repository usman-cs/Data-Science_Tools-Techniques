##### Probability of mushroom to be poissonous
$$
P(class ='p') = \frac {num\;of \;mushroom \;with \;p}{total \;num \;of \;sample}
$$


```python
print('Answer: ',df[df['class']=='p'].shape[0]/df.shape[0])
```

\begin{align}
P(cap-shape = x \cap cap-surface = y \cap cap-color = w | class = p) = \frac{P(cap-shape = x \cap cap-surface = y \cap cap-color \cap class = p)}{P(class = p)}
\end{align}

```python
print("Answer: ", (df[(df['cap-shape']=='x') & (df['cap-color']=='w') & (df['cap-surface']=='y') & (df['class']=='p')].shape[0] / df.shape[0]) / (df[df['class']=='p'].shape[0]/df.shape[0])) 
```
